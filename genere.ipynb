```python
import random
import pandas as pd

# Ce script génère un dataset de profils utilisateurs pour une plateforme de streaming de films, avec des anomalies intentionnelles pour tester les capacités de nettoyage et d'analyse des données.
tous_les_noms = ["Liam", "Noah", "Oliver", "Elijah", "James", "William", "Benjamin", "Lucas","Emma", "Olivia", "Ava", "Isabella", "Sophia", "Mia", "Charlotte", "Amelia"]
# Liste des genres de films disponibles
genres_disponibles = ["action", "thriller", "drame", "sci-fi", "horreur", "romance", "comédie"]
# Banque de films classés par genre, avec des titres et des réalisateurs fictifs
banque_films = {
    "action": [
        {"movie": "Mad Max: Fury Road", "director": "George Miller"},
        {"movie": "John Wick", "director": "Chad Stahelski"},
        {"movie": "The Dark Knight", "director": "Christopher Nolan"}
    ],
    "thriller": [
        {"movie": "Se7en", "director": "David Fincher"},
        {"movie": "Shutter Island", "director": "Martin Scorsese"},
        {"movie": "Gone Girl", "director": "David Fincher"}
    ],
    "drame": [
        {"movie": "The Shawshank Redemption", "director": "Frank Darabont"},
        {"movie": "Forrest Gump", "director": "Robert Zemeckis"},
        {"movie": "Fight Club", "director": "David Fincher"}
    ],
    "sci-fi": [
        {"movie": "Inception", "director": "Christopher Nolan"},
        {"movie": "Interstellar", "director": "Christopher Nolan"},
        {"movie": "The Matrix", "director": "Lana Wachowski"}
    ],
    "horreur": [
        {"movie": "The Conjuring", "director": "James Wan"},
        {"movie": "Get Out", "director": "Jordan Peele"}
    ],
    "romance": [
        {"movie": "The Notebook", "director": "Nick Cassavetes"},
        {"movie": "La La Land", "director": "Damien Chazelle"}
    ],
    "comédie": [
        {"movie": "The Hangover", "director": "Todd Phillips"},
        {"movie": "Superbad", "director": "Greg Mottola"}
    ]
}

# Fonction pour générer des données de simulation avec des anomalies intégrées
def generer_donnees_simulation(nombre_utilisateurs):
    liste_profils = []
    
    for _ in range(nombre_utilisateurs):
        nom = random.choice(tous_les_noms)
        age = random.randint(15, 65)
        
        nb_genres = random.randint(1, 3)
        preferences = random.sample(genres_disponibles, nb_genres)
        
        historique_visionnage = []
        nb_films = random.randint(1, 4)
        
        for _ in range(nb_films):
            if random.random() < 0.80:
                genre_choisi = random.choice(preferences)
            else:
                genre_choisi = random.choice(genres_disponibles)
                
            film_modele = random.choice(banque_films[genre_choisi])
            
            # --- INCORPORATION D'ANOMALIES DANS L'HISTORIQUE ---
            note_attribuee = random.randint(3, 5)
            
            # 1. Donnée aberrante sur la note : 2% de chances d'avoir une note impossible (ex: -1 ou 99)
            if random.random() < 0.02:
                note_attribuee = random.choice([-1, 99])
                
            film_historique = {
                "movie": film_modele["movie"],
                "genre": genre_choisi,
                "rating": note_attribuee
            }
            
            # 2. Donnée manquante dans l'historique 
            if random.random() < 0.03:
                film_historique["genre"] = None
            
            if film_historique not in historique_visionnage:
                historique_visionnage.append(film_historique)
                
        # --- INCORPORATION D'ANOMALIES DANS LE PROFIL ---
        
        # 3. Donnée manquante sur l'âge :
        if random.random() < 0.05:
            age = None
            
        # 4. Donnée aberrante sur l'âge
        elif random.random() < 0.02:
            age = random.choice([-5, 150])
            
        # 5. Donnée manquante sur les préférences
        if random.random() < 0.04:
            preferences = []

        profil = {
            "name": nom,
            "age": age,
            "preferences": preferences,
            "watch_history": historique_visionnage
        }
        liste_profils.append(profil)
        
    return liste_profils


# 1. DEFINITION DE LA QUANTITE DE DONNEES

NOMBRE_DE_LIGNES = 1500

donnees_brutes = generer_donnees_simulation(NOMBRE_DE_LIGNES)


# 2. CREATION DU DATAFRAME PANDAS

df = pd.DataFrame(donnees_brutes)

print(f"Structure vérifiée pour {len(df)} utilisateurs générés.")
df.head()


# 3. CONVERSION ET EXPORT EN CSV

df.to_csv("./data/profils_utilisateurs.csv", index=False, encoding="utf-8")
 
```
